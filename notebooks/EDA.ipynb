A continuación crearemos 3 graficos que nos proporcionaran información importante sobre nuestro DataFrame

```python
# Importamos las librerias que se utilizaran
import pandas as pd
import plotly.express as px
```

```python
# Imortamos nuestro archivo csv
car_data = pd.read_csv(r"C:\Users\sebas\Desktop\gits\Tripleten\proyecto_sprint_7\vehicles_us.csv")
```

# Histograma 

Muestra la distribución del kilometraje de los vehículos y permite identificar qué rango de uso es más común en el conjunto de datos.

```python
# Creamos nuestro histograma

fig = px.histogram(car_data, x="odometer") # crear un histograma
fig.show() # crear gráfico de dispersión
```

# Grafico de dispersion

Permite analizar la relación entre kilometraje y precio, mostrando cómo el valor del vehículo cambia según su uso y condición.

```python
# Creamos un scatter plot 
px.scatter(car_data, x="odometer", y="price", color="condition")
```

# Diagrama de caja

Compara la distribución de precios según la condición del vehículo, destacando diferencias en medianas y dispersión entre categorías.

```python
# Creamos un boxplot
px.box(car_data, x="condition", y="price")
```
